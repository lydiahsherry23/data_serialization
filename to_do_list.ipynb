```python
tasks = [
        "eat lunch",
        "complete assignment for submission",
        "do chores",
        "go to the market",
        "read a novel"
    ]
def tasks_to_do(tasks):
    """Add a new task to the list."""
    task_description = input("\nEnter the task description: ").strip()
    if task_description:
        tasks.append((task_description, False))
        print(f"Task '{task_description}' added.")
    else:
        print("Task description cannot be empty.")
tasks_to_do(tasks)
```

```python
def to_do_list_menu():
    """Display to_do_list_menu."""
    print("\nTo-Do List Menu")
    print("1. Add Task")
    print("2. Mark Task as Completed")
    print("3. List Tasks")
    print("4. Remove Task")
    print("5. Quit")

def tasks_to_do(tasks):
    """Display all tasks whether done or not done."""
    if not tasks:
        print("\nNo tasks in the list.")
    else:
        print("\nTasks:")
        for index, (description, done) in enumerate(tasks, start=1):
            status = "done" if done else "not done"
            print(f"{index}. {description} [{status}]")

def add_task(tasks):

    task_description = input("\nEnter the task description: ").strip()
    if task_description:
        tasks.append((task_description, False))
        print(f"Task '{task_description}' added.")
    else:
        print("Task description cannot be empty.")

def tasks_completed(tasks):
    """Mark a task as done."""
    tasks_to_do(tasks)
    if tasks:
        try:
            task_num = int(input("\nEnter the task number to mark as done: "))
            if 1 <= task_num <= len(tasks):
                task, done = tasks[task_num - 1]
                if done:
                    print(f"Task '{task}' is already marked as done.")
                else:
                    tasks[task_num - 1] = (task, True)
                    print(f"Task '{task}' marked as done.")
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")

def remove_task(tasks):

    tasks_to_do(tasks)
    if tasks:
        try:
            task_num = int(input("\nEnter the task number to remove: "))
            if 1 <= task_num <= len(tasks):
                removed_task, _ = tasks.pop(task_num - 1)
                print(f"Task '{removed_task}' removed.")
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")

def main():
    """Main function to run the to-do list manager."""
    tasks = []
    while True:
        to_do_list_menu()
        choice = input("\nChoose an option: ").strip()
        if choice == '1':
            add_task(tasks)
        elif choice == '2':
            tasks_completed(tasks)
        elif choice == '3':
            tasks_to_do(tasks)
        elif choice == '4':
            remove_task(tasks)
        elif choice == '5':
            print("that's a trial")
            break
        else:
            print("Not among the choices. Please choose one of the above choices.")

if __name__ == "__main__":
    main()
```
